```python
import numpy as np
import matplotlib.pyplot as plt

# Create a range of x values for the semicircle
x = np.linspace(-10, 10, 1000)

# Calculate corresponding y values for the upper semicircle
y_upper = np.sqrt(100 - x**2)

# Calculate corresponding y values for the lower semicircle
y_lower = np.sqrt(9 - x**2)

# Create the figure and axis
fig, ax = plt.subplots()

# Plot the upper semicircle
ax.plot(x, y_upper, label='Upper Semicircle', color='blue')

# Plot the lower semicircle
ax.plot(x, y_lower, label='Lower Semicircle', color='red')

# Fill the area between the semicircles
ax.fill_between(x, y_upper, y_lower, color='gray', alpha=0.5, label='Filled Area')

# Set axis labels and a legend
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()

# Set the aspect ratio to be equal
ax.set_aspect('equal')

# Show the plot
plt.title('Fill between Semicircles: x^2 + y^2 = 100 and x^2 + y^2 = 9')
plt.grid(True)
plt.show()
```
